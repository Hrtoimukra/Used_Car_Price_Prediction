```python
def extract_numeric(value):
    value=value.split(' ')[0].strip()
    if value=='':
        return 0.0
    return float(value)   
```

```python
def get_brand_name(car_model):
    return car_model.split(' ')[0].strip()
```

```python
cars_data
```

```python
cars_data['mileage']=cars_data['mileage'].apply(extract_numeric)
```

```python
cars_data['engine']=cars_data['engine'].apply(extract_numeric)
```

```python
cars_data['max_power']=cars_data['max_power'].apply(extract_numeric)
```

```python
cars_data['name']=cars_data['name'].apply(get_brand_name)
```

```python
cars_data['seats']=cars_data['seats'].astype(int)
```

```python
predict=model.predict(X_test)
```
